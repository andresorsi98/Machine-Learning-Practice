```python
import numpy as np
import pandas as pd
```

```python
dataSet = pd.read_csv("housing_cleansed.csv")
```

```python
X_cols = list( set(dataSet.columns) - set(['price']) )
y_cols = ['price']

X = dataSet[X_cols].values
y = dataSet[y_cols].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import sklearn.metrics as metrics
```

```python
#Multiple Linear Regression
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

y_pred_mlr = linear_regressor.predict(X_test)


r2_mlr = metrics.r2_score(y_test, y_pred_mlr)
print("R2 MLR: ", r2_mlr)
```

```python
#Ridge
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)

r2_ridge = metrics.r2_score(y_test, y_pred_ridge)
print("R2 Ridge: ", r2_ridge)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha' : [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
}

ridge_cv = GridSearchCV(ridge, param_grid, cv = 3, n_jobs = -1)
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
y_pred2 = ridge_cv.predict(X_test)
```

```python
r2_2 = metrics.r2_score(y_test, y_pred2)
```

```python
print("R2_2: ", r2_2)
```

```python
#Lasso Model
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1.0, max_iter=100000)
lasso.fit(X_train, y_train)

y_pred_lasso = lasso.predict(X_test)
```

```python
r2_lasso = metrics.r2_score(y_test, y_pred_lasso)
print("R2 Lasso: ", r2_lasso)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'alpha' : [0.01, 0.1, 1, 10, 1000]
}
```

```python
lasso_cv = GridSearchCV(lasso, param_grid, cv = 3, n_jobs = 1)
lasso_cv.fit(X_train, y_train)
```

```python
y_pred_lassocv = lasso_cv.predict(X_test)

r2_lassGridSearch = metrics.r2_score(y_test, y_pred_lassocv)
print("R2 Lasso GridSearch: ", r2_lassGridSearch)
```

```python
bestLasso = Lasso(alpha = 1000, max_iter = 100000)
bestLasso.fit(X_train, y_train)
```

```python
y_pred_bl = bestLasso.predict(X_test)
```

```python
r2_bestLasso = metrics.r2_score(y_test, y_pred_bl)

print("R2 Best Lasso: ", r2_bestLasso)
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

forestRegressor = RandomForestRegressor(random_state = 13)
```

```python
forestRegressor.fit(X_train, y_train)
```

```python
y_pred_rfr = forestRegressor.predict(X_test)
```

```python
r2_rfr = metrics.r2_score(y_test, y_pred_rfr)

print("R2 Score: ", r2_rfr)
```
